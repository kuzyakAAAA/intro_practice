```python
import requests
import pandas as pd  

LOGINOM_URL = "https://edu.loginom.dev/lgi/rest/instacart_ws_kuzmin1/GetUserHistory"
USER_ID = 7
payload = {
    "Variables": {
    "user_id": USER_ID
}
}
```

```python
response = requests.post(LOGINOM_URL, json=payload, timeout=30)
response.raise_for_status()
rows = response.json()["DataSet"]["Rows"]
## если необходимо преобразовать ответ сервиса в датафрейм, используйте конструкцию ниже
df= pd.DataFrame(rows)
df.head(10)
```

```python
lines = []
for i, row in enumerate(rows, start=1):
    lines.append(
        f"{i}. {row['product_name']} —"
        f"отдел: {row['department_rus']}, "
        f"категория: {row['aisle']}, "
        f"заказов: {row['order_count']}"
    )
products_text= "\n".join(lines)
print(products_text)
```
